```python
def distance(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295     #Pi/180
    a = 0.5 - cos((lat2 - lat1) * p)/2 + cos(lat1 * p) * cos(lat2 * p) * (1 - cos((lon2 - lon1) * p)) / 2
    return 12742 * asin(sqrt(a))
```

```python
for i in range(len(df1)):
    aux={'dist':[],'lat_fin':[],'lon_fin':[]}
    lat = df1.loc[i,'vlr_latitude']
    lon = df1.loc[i,'vlr_longitude']
    for j in range(len(df1)):
        if j!=i:
            aux['dist'].append(distance(lat,lon,df1.loc[j,'vlr_latitude'],df1.loc[j,'vlr_longitude']))
            aux['lat_fin'].append(df1.loc[j,'vlr_latitude'])
            aux['lon_fin'].append(df1.loc[j,'vlr_longitude'])
    
    aux = pd.DataFrame(aux)
    aux_arr= np.array(aux['dist'])
    df1['min_distance'] = aux_arr.min()
    df1['lat_fin'] = aux.iloc[aux_arr.argmin()].values[1]
    df1['lon_fin'] = aux.iloc[aux_arr.argmin()].values[1]
```
